# titanic dataset

### titanic dataset

PassengerId: A unique identifier for each passenger (integer). 

Pclass:  Passenger class(1 = 1st, 2 = 2nd, 3 = 3rd).  This is a categorical feature representing socio-economic status.

Name:  Name of the passenger (string).  While it might seem irrelevant, the titles in the name (Mr., Mrs., Miss, etc.)

Sex:  Sex of the passenger (string: Male or Female).  This is a categorical feature.

Age:  Age of the passenger (float or integer). This is a numerical feature, but it has missing values.

SibSp:  Number of siblings/spouses aboard the Titanic (integer).  This is a numerical feature.

Parch:  Number of parents/children aboard the Titanic (integer).  This is a numerical feature.

Ticket:  Ticket number (string). This is mostly a unique identifier and is often not useful in basic analysis, though the ticket prefix might reveal some group information.

Fare:  Passenger fare (float). This is a numerical feature, representing the cost of the ticket.

Cabin:  Cabin number (string).  This is a categorical feature, but it has a lot of missing values. The cabin letter might be useful for feature engineering.

Embarked:  Port of embarkation (string: C = Cherbourg, Q = Queenstown, S = Southampton). This is a categorical feature.

Survived:  Indicates whether the passenger survived (0 = No, 1 = Yes).

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('Titanic.csv')
```

```python
plt.figure(figsize=(8,6))
sns.histplot(titanic["Age"].dropna(), bins=30, kde=True, color='blue')
plt.title("histogram of passenger age",fontsize=14)
plt.xlabel("Age", fontsize=14)
plt.ylabel("frequency", fontsize=14)
plt.grid(True, linestyle='-.', alpha=0.8) 
plt.savefig("histogram.png", dpi=300, bbox_inches="tight")#dpi:dots per inch-decide the resolution (quality)
plt.show()                                                #bbox_box:removes extra space around the plot
```

# 2. Box Plot (Fare)

```python
# 2. Box Plot (Fare) #to see the distribution of a single numerical variable 
plt.figure(figsize=(6, 4))
sns.boxplot(x=titanic['Fare'])
plt.title('Distribution of Passenger Fares')
plt.show()
```

# 3. Density Plot (Age)

```python

plt.figure(figsize=(8, 6))
sns.kdeplot(titanic['Age'].dropna())
plt.title('Density Plot of Passenger Ages')
plt.show()
```

```python
# Plot KDE (Density Plot)
plt.figure(figsize=(8, 6))
sns.kdeplot(titanic['Age'].dropna(), fill=True, color="blue", alpha=0.5)  # `shade=True` is deprecated, use `fill=True`
plt.title('Density Plot of Passenger Ages', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# 4. Scatter Plot (Age vs. Fare)

```python
# 
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic, hue='Survived')
plt.title('Age vs. Fare (Colored by Survival)')
plt.show()

#hue=It tells Seaborn to color each point on the plot based on the value in the Survived column.
#This adds a third dimension of information
```

# 5. Correlation Heatmap

```python
#this is called as pearson's corelation.>>>-1 to +1  >>>
# Select numerical columns
numerical_cols = titanic.select_dtypes(include=['number'])  # Or include=['int64', 'float64'] for more specificity

# Calculate the correlation matrix using only the numerical columns
correlation_matrix = numerical_cols.corr()

# Now lets draw heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues') #writes the numerical value of each cell directly onto the heatmap
plt.title('Correlation Matrix of Numerical Features')
plt.show()

#'viridis' 'plasma' 'magma' 'Blues' 'Greens'
```

# 6.Bar plot (or count plot) for Sex

```python

plt.figure(figsize=(6, 4))  # Adjust figure size if needed
sns.countplot(x='Sex', data=titanic) # Or sns.barplot(x=titanic['Sex'].value_counts().index, y=titanic['Sex'].value_counts().values)
plt.title('Distribution of Passenger Sex')
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.show()
```

```python

plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

## 7. Pie Chart (Survived)

```python


plt.figure(figsize=(6, 6))
titanic['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%') 
plt.title('Overall Survival Rate')
plt.show()
#autopct is a parameter that controls how the percentage of each slice is displayed on the chart. 
```

## 8. Violin Plot (Age by Pclass)

>> to visualize the distribution of numerical data across different categories. 
>> It's a hybrid of a box plot and a Kernel Density Estimate (KDE) plot.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Pclass', y='Age', data=titanic, hue='Survived', split=True)
plt.title('Age Distribution by Pclass and Survival')
plt.show()

#Like a box plot, it shows summary statistics: the white dot in the middle is the median, 
# and the thick black bar is the interquartile range.

# Like a KDE plot, the width of the violin shows the frequency (or density) of data at different values. 
# A wider section means more data points are concentrated there.
```

## 9. Pair Plot (Numerical Features)

```python

numerical_titanic = titanic[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
sns.pairplot(numerical_titanic, hue='Survived')
plt.show()
```

## line plot

A line plot is used to show a trend or the change in a numerical value over a continuous and ordered interval, which is most often time. It connects a series of data points with a continuous line, making it excellent for visualizing trends, patterns, and seasonality.

```python
sales_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales_in_Lakhs': [150, 180, 220, 190, 250, 270]
}
df_sales = pd.DataFrame(sales_data)
```

```python
df_sales
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="Month", y="Sales_in_Lakhs", data=df_sales, marker='o', sort=False )
plt.title("monthly sales trend for first half of the year")
plt.xlabel("month",fontsize=12)
plt.ylabel("sales(in lakhs)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.savefig("lineplot.pdf", dpi=300, bbox_inches="tight") #.jpg, .png, .pdf
plt.show()
```
