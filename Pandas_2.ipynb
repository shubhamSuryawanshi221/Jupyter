## Select data based on some condition

```python


survivors = df[df['survived'] == 1]

print("\nPassengers who Survived:")
print(survivors.head())
survivors
```

```python
#all female passengers who were in 1st class. sex>>0 and pclass>>>1

# Note the parentheses around each condition
female_first_class = df[(df['sex'] == 0) & (df['pclass'] == 1)] #and

print("Female passengers from 1st Class:")
print(female_first_class.head())
female_first_class
```

```python
#all passengers who either survived OR were older than 60. 1>>>survived.....age>>>60

# Note the parentheses around each condition
survived_or_senior = df[(df['survived'] == 1) | (df['age'] > 60)] #or 

print("Passengers who Survived OR were older than 60:")
print(survived_or_senior.tail())
survived_or_senior
```
